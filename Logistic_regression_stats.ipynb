```python
features = [
    "Q3. How much did you pay (in INR) for your smartwatch? Enter the amount in Indian Rupees (₹).",
    "Q10. How long (in hours) does the battery last on average per full charge? ( Enter the number of hours. )",
    "Q8. How many hours per day (on average) do you wear the smartwatch? ( Enter the number of hours.)",
    "Q9. How many times per day do you interact with or check the watch? ( Enter the number of times per day )",
    "Q16. Rate your willingness to recommend this smartwatch to others (0 = Not at all, 10 = Definitely)"
]
```

```python
df_clean = df.dropna(subset=features + ["is_satisfied"])
X = df_clean[features]
y = df_clean["is_satisfied"]
```

To evaluate model performance, we split data into:
- Training set → used to fit the model
- Test set → used to validate the model on unseen data
stratify=y ensures both classes (0 and 1) are proportionally represented.

We’re using logistic regression to predict satisfaction.
- liblinear: works well on small datasets
- class_weight="balanced": compensates for possible class imbalance

These metrics help us judge how well our model performs:
- Precision, Recall, F1-score → classification accuracy
- AUC Score → measures the model’s ability to rank satisfied vs. not satisfied users across thresholds


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

X = df_clean[features]
y = df_clean["is_satisfied"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

model = LogisticRegression(solver="liblinear", class_weight="balanced")
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("AUC score:", roc_auc_score(y_test, y_proba))
```

```python
import statsmodels.api as sm

X_with_const = sm.add_constant(X_train)  # adds intercept
ols_model = sm.OLS(y_train, X_with_const).fit()
print(ols_model.summary())
```
